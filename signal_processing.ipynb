```python
#Signal Processing Simulations in Python
import numpy as np
import scipy
from scipy import signal
import matplotlib.pyplot as plt
```

```python
def setplot(xmin,xmax,ymin,ymax,figsize=[18,10]):  #Custom wrapper for plot config
    axes = plt.gca()
    axes.set_xlim([xmin,xmax])
    axes.set_ylim([ymin,ymax])
    fig=plt.gcf()
    fig.set_size_inches(figsize[0],figsize[1])

    
#Define Filters/Signal Operations

def filter_clip(signal,threshold=0.5):  #Clipping Filter
    for s in range(0,signal.size):
        if signal[s]>threshold: 
            signal[s]=threshold
        elif signal[s]<(-threshold):
            signal[s]=-threshold
    return signal   

def filter_invert(signal):  #Signal Inversion Filter
    return signal
```

```python
#Generate some signal waveforms
tmax=1
pi=np.pi
t=np.linspace(0,tmax,500) #sampled at 500 Hz for 1 second
s1=np.sin(2*pi*6*t)
plt.plot(t,s1)

#Apply Filters/Signal Operations
fs1=filter_clip(s1,0.6)

#Plot Results
setplot(0,tmax,-1,1,[5,5])
plt.plot(t,fs1)
```

```python
#Non-periodic signal
tmax=10
t=np.arange(0,tmax,0.01)
s2=3*np.sinc((t-5)*5)
setplot(0,tmax,-1,1,[5,5])
plt.plot(t,s2)
plt.plot(t,filter_clip(s2,0.5))
```

```python
tmax=1
t=np.linspace(0,tmax,500) #sampled at 500 Hz for 1 second
s3=signal.sawtooth(2*pi*5*t)
setplot(0,tmax,-1,1,[5,5])
plt.plot(t, s3)
```

```python
#PWM Test

# sig = np.sin(2 * np.pi * t)
sig=np.sin(2*pi*6*t)
pwm = signal.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
setplot(0,1,-1.5,1.5,[8,8])
plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.subplot(2, 1, 2)
plt.plot(t, pwm)
```
